```python
def impact(X):
    if X >=80:
        return 0    # 0: 'Very High'
    elif X >= 60:
        return 1    # 1: 'High'
    elif X >= 40:
        return 2    # 2: 'Moderate'
    elif X >= 20:
        return 3    # 3: 'Low'
    else:
        return 4    # 4: 'Very Low'
```

```python
df['ResultTarget']=df['HealthImpactScore'].apply(impact)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['HealthImpactScore','HealthImpactClass', 'ResultTarget'],axis=1)
y=df['ResultTarget']
```

```python
y=np.array(y)
```

```python
model_names = []
accuracies = []
```

**TRAINING**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=58)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

**LogisticRegression**

```python
from sklearn.linear_model import LogisticRegression

log_reg_model = LogisticRegression()
log_reg_model.fit(X_train, y_train)

log_reg_accuracy = log_reg_model.score(X_test, y_test)
print(f"Lojistik Regresyon Modeli Doğruluğu: {log_reg_accuracy:.5f}")

model_names.append('LogisticRegression')
accuracies.append(log_reg_accuracy)
```

```python
y_pred = log_reg_model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

plt.figure(figsize=(9, 7))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Reds')
plt.title('Confusion Matrix', fontsize='16')
plt.xlabel('Predicted', fontsize='14')
plt.ylabel('True', fontsize='14')
plt.show()
```

**ArtificialNeuralNetwork**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
y_train = encoder.fit_transform(y_train.reshape(-1, 1)).toarray()
y_test = encoder.transform(y_test.reshape(-1, 1)).toarray()
```

```python
ann_model = Sequential()
ann_model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
ann_model.add(Dense(64, activation='relu'))
ann_model.add(Dense(64, activation='relu'))
ann_model.add(Dense(4, activation='softmax'))
ann_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
ann_model.fit(X_train, y_train, epochs=30,batch_size=32,validation_data=(X_test, y_test), verbose=0)
#ann_model.fit(X_train, y_train, epochs=30,batch_size=32,validation_data=(X_test, y_test), verbose=2)
```

```python
loss=pd.DataFrame(ann_model.history.history)
loss
```

```python
plt.figure(figsize=(9,7))
sns.lineplot(loss)
```

```python
ann_predictions=(ann_model.predict(X_test)>0.5).astype('int32')
```

```python
print(classification_report(y_test,ann_predictions))
```

```python
losses,ann_accuracy=ann_model.evaluate(X_test,y_test)
print(f"Test Loss Is: {round(losses,4)},\nTest Accuracy Is: {round(ann_accuracy,4)}")
```

```python
new_row = pd.DataFrame({'Model': ['ArtificialNeuralNetwork'], 'Accuracy': [ann_accuracy]})
results_df = pd.concat([results_df, new_row], ignore_index=True)
```

```python
results_df.sort_values(by='Accuracy', ascending=False)
```
